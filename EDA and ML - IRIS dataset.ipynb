```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score

X = iris.drop('Name', axis = 1)
y = iris['Name'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

result = model.predict(X_test)
```

```python
print(classification_report(y_test, result))
```

```python
accuracy_score(y_test, result)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)

result = model.predict(X_test)
print(classification_report(y_test, result))
```

```python
accuracy_score(y_test, result)
```

```python
knn.score(X_test, y_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

model = GaussianNB()
model.fit(X_train, y_train)

result = model.predict(X_test)
print(classification_report(y_test, result))
```

```python
accuracy_score(y_test, result)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.preprocessing import LabelEncoder

iris_km = iris.copy()

encoder = LabelEncoder()

iris_km['Name'] = encoder.fit_transform(iris_km['Name'])
iris_km
```

```python
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import classification_report, accuracy_score

X = iris_km.drop('Name', axis = 1)
y = iris_km['Name'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

km = KMeans(n_clusters = 3, max_iter = 1000)
km.fit(X_train, y_train)

train_labels = km.predict(X_train)
test_labels = km.predict(X_test)
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
accuracy_score(y_train, train_labels)
```

```python
accuracy_score(test_labels, y_test)
```

```python
print(classification_report(y_train, train_labels))
```
